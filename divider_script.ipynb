```
import numpy as np
n = int(input("enter the value of N for 2N/N divider:"))
entries = list(map(int, input("Enter a NxN matrix with 0-> exact and 1-> approximate").split())) 
matrix = np.array(entries).reshape(n, n) 
print(matrix)


divider = open("divider_script.v","w")

divider.write("module divider_script(n,d,q,r);"+"\n")
divider.write("input ["+str(2*(n)-1)+":0]n;"+"\n")
divider.write("input ["+str(n-1)+":0]d;"+"\n")
divider.write("input ["+str(n-1)+":0]q,r;"+"\n \n")

divider.write("wire ["+str(n-1)+":0]bout_local[0:"+str(n-1)+"];"+"\n")
divider.write("wire ["+str(n-1)+":0]r_local[0:"+str(n-1)+"];"+"\n \n")
x=0
for i in range(n):
  if matrix[n-1-i][n-1] == 1:
    divider.write("subtractor_approx sb"+str(x)+"(n["+str(i)+"],d[0],0,q["+str(i)+"],r_local["+str(i)+"][0],bout_local["+str(i)+"][0]);"+"\n");
  else:
    divider.write("subtractor sb"+str(x)+"(n["+str(i)+"],d[0],0,q["+str(i)+"],r_local["+str(i)+"][0],bout_local["+str(i)+"][0]);"+"\n");
  i=i+1
  x=x+1

i=0
for i in range(n-1):
  if matrix[0][n-2-i] == 1:
    divider.write("subtractor_approx sb"+str(x)+"(n["+str(x)+"],d["+str(i+1)+"],bout_local["+str(n-1)+"]["+str(i)+"]"+",q["+str(n-1)+"],r_local["+str(n-1)+"]["+str(i+1)+"],bout_local["+str(n-1)+"]["+str(i+1)+"]);"+"\n");
  else:
    divider.write("subtractor sb"+str(x)+"(n["+str(x)+"],d["+str(i+1)+"],bout_local["+str(n-1)+"]["+str(i)+"]"+",q["+str(n-1)+"],r_local["+str(n-1)+"]["+str(i+1)+"],bout_local["+str(n-1)+"]["+str(i+1)+"]);"+"\n");
  i=i+1
  x=x+1

i=0
for i in range(n-1):
  for j in range(n-1):
    if matrix[n-i-1][n-2-j] == 1:
      divider.write("subtractor_approx sb"+str(x)+"(r_local["+str(i+1)+"]["+str(j)+"],d["+str(j+1)+"],bout_local["+str(i)+"]["+str(j)+"]"+",q["+str(i)+"],r_local["+str(i)+"]["+str(j+1)+"],bout_local["+str(i)+"]["+str(j+1)+"]);"+"\n");
    else:
      divider.write("subtractor sb"+str(x)+"(r_local["+str(i+1)+"]["+str(j)+"],d["+str(j+1)+"],bout_local["+str(i)+"]["+str(j)+"]"+",q["+str(i)+"],r_local["+str(i)+"]["+str(j+1)+"],bout_local["+str(i)+"]["+str(j+1)+"]);"+"\n");
    j=j+1
    x=x+1
  i=i+1

i=0
for i in range(n-1):
  divider.write("assign q["+str(i)+"] = (r_local["+str(i+1)+"]["+str(n-1)+"])|(~bout_local["+str(i)+"]["+str(n-1)+"]); \n")

divider.write("assign q["+str(n-1)+"] = (n["+str((2*n-1))+"])|(~bout_local["+str(n-1)+"]["+str(n-1)+"]); \n")

i =0
for i in range(n):
  divider.write("assign r["+str(i)+"] = r_local[0]["+str(i)+"]; \n")

divider.write("endmodule \n")
divider.close()



```
